```python
pulsos_num = lencoder.fit_transform(dados_treino["PULSOS"])
ppa_num = lencoder.fit_transform(dados_treino["PPA"])
b2_num = lencoder.fit_transform(dados_treino["B2"])
sopro_num = lencoder.fit_transform(dados_treino["SOPRO"])
hda1_num = lencoder.fit_transform(dados_treino["HDA 1"])
hda2_num = lencoder.fit_transform(dados_treino["HDA2"])
sexo_num = lencoder.fit_transform(dados_treino["SEXO"])
motivo1_num = lencoder.fit_transform(dados_treino["MOTIVO1"])
motivo2_num = lencoder.fit_transform(dados_treino["MOTIVO2"])
classe_num = lencoder.fit_transform(dados_treino["CLASSE"])
fc_num = lencoder.fit_transform(dados_treino['FC'])
```

```python
dados_treino["PULSOS"] = pulsos_num
dados_treino["PPA"] = ppa_num
dados_treino["B2"] = b2_num
dados_treino["SOPRO"] = sopro_num
dados_treino["HDA 1"] = hda1_num
dados_treino["HDA2"] = hda2_num
dados_treino["SEXO"] = sexo_num
dados_treino["MOTIVO1"] = motivo1_num
dados_treino["MOTIVO2"] = motivo2_num
dados_treino["CLASSE"] = classe_num
dados_treino['FC'] = fc_num
```

```python
dados_teste["PULSOS"] = pulsos_num_teste
dados_teste["PPA"] = ppa_num_teste
dados_teste["B2"] = b2_num_teste
dados_teste["SOPRO"] = sopro_num_teste
dados_teste["HDA 1"] = hda1_num_teste
dados_teste["HDA2"] = hda2_num_teste
dados_teste["SEXO"] = sexo_num_teste
dados_teste["MOTIVO1"] = motivo1_num_teste
dados_teste["MOTIVO2"] = motivo2_num_teste
dados_teste["FC"] = fc_num_test
```

```python
dados_teste = dados_teste.drop(columns=["DN", "Atendimento", 'Convenio'])
```

```python
display(dados_teste)
```

```python
dados_teste["IDADE"] = pd.to_numeric(dados_teste["IDADE"], errors="coerce")
```

```python
dados_teste = dados_teste.fillna(dados_teste.median())
```

```python
atributos_test = [
    "Peso",
    "Altura",
    "IMC",
    "IDADE",
    "PULSOS",
    "PA SISTOLICA",
    "PA DIASTOLICA",
    "PPA",
    "B2",
    "SOPRO",
    "FC",
    "HDA 1",
    "HDA2",
    "SEXO",
    "MOTIVO1",
    "MOTIVO2",
]

atributos = [
    "Peso",
    "Altura",
    "IMC",
    "IDADE",
    "PULSOS",
    "PA SISTOLICA",
    "PA DIASTOLICA",
    "PPA",
    "B2",
    "SOPRO",
    "FC",
    "HDA 1",
    "HDA2",
    "SEXO",
    "MOTIVO1",
    "MOTIVO2",
]

target = ['CLASSE']

X_train = dados_treino[atributos]
X_test = dados_teste[atributos_test]
y_train = dados_treino[target]
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
modelo = MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000)
```

```python
modelo.fit(X_train, y_train)
```

```python
X_test = dados_teste[atributos_test]

y_pred = modelo.predict(X_test)

print(y_pred)


```

```python
df = pd.DataFrame(dados_teste["Id"], columns=["Id"])
df["Predicted"] = y_pred
display(df)

df.to_csv('predicao.csv', index=False)
```
